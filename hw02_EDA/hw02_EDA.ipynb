### B) Пропуски и очистка

Исходя из быстрого обзора данных, пропуски есть в колонках: 
**name** *(заголовок объявления о сдаче апартаментов)*, 
**host_name** *(имя владельца)*, 
**last_review** *(дата последнего отзыва)*,
**reviews_per_month** *(среднее кол-во отзывов в месяц)*.

#### Признак **name**

Колонка содержит заголовки объявлений об аренде апартаментов. В нынешнем виде такая информация почти полностью будет бесполезна для обучения моделей, если мы, например, захотим предсказывать цену апартаментов. Поэтому лучшим решением будет **удалить** колонку полностью.

```python
df_clean = df.copy()

df_clean = df_clean.drop("name", axis=1)
```

#### Признак **host_name**

Колонка содержит совершенно обычную строковую информацию - имя владельца апартаментов. Учитывая наличие признака **host_id**, по которому явно будет лучше идентифицировать владельца, данную колонку тоже стоит **удалить** из датасета.

```python
df_clean = df_clean.drop("host_name", axis=1)
```

#### Признак **last_review** и **reviews_per_month** 

Колонка **last_review** содержит дату последнего отзыва в формате строки. Пропуски в данном случае означают то, что у объявления попросту нет пока отзывов. Проверим эту гипотезу с помощью колонки **number_of_reviews** - если окажется, что при значении 0 в следующих двух колонках отсутствует значение, то объявление действительно пока не имеет отзывов:

```python
df[df['number_of_reviews'] == 0][['last_review', 'reviews_per_month']].isna().mean()
```

Так и оказалось! Теперь можно принять решение относительно пропусков. Так, сама по себе дата последнего отзыва для потенциальной модели мало что значит, в то время как общее количество отзывов и их количество в месяц показывает, насколько популярны аппартаменты. Поэтому оптимальным решением будет удалить колонку **last_review**, а колонку **reviews_per_month** в месте пропусков заполнить значением 0.

```python
df_clean = df_clean.drop("last_review", axis=1)
```

```python
df_clean.reviews_per_month = df_clean.reviews_per_month.fillna(0)
```

### Оne-hot encoding

Категориальные признаки **neighbourhood_group** и **room_type** имеют не очень большое количество категорий - 5 и 3 соответственно. Поэтому для кодирования подойдёт **one-hot encoding**.

```python
df_encoded = pd.get_dummies(df_clean, columns=["neighbourhood_group", "room_type"])
df_encoded.iloc[:, 10:].head()
```

### Frequency encoding и Target encoding

Категориальный признак **neighbourhood** отличается большим количеством уникальных значений - 221, что делает использование того же one-hot encoding очень нецелесообразным *(получится очень большая матрица, состоящая в основном из нулей)*. Поэтому подойдём к кодированию признака с другой стороны. Рассмотрим два способа: по частоте встречаемости в колонке **(frequency encoding)** и по средней цене в районе **(target encoding)**.

#### Frequency encoding

```python
df_encoded_freq = df_encoded.copy()
```

```python
freq = df_encoded_freq["neighbourhood"].value_counts()
df_encoded_freq["neighbourhood_freq"] = df_encoded_freq["neighbourhood"].map(freq)
df_encoded_freq = df_encoded_freq.drop("neighbourhood", axis=1)
df_encoded_freq.head()
```

#### Target encoding

```python
df_encoded_target = df_encoded.copy()
```

```python
neighbourhoods_mean_price = df_encoded_target.groupby("neighbourhood")["price"].mean()
df_encoded_target["neighbourhood_target"] = df_encoded_target["neighbourhood"].map(neighbourhoods_mean_price)
df_encoded_target = df_encoded_target.drop("neighbourhood", axis=1)
df_encoded_target.head()
```

Несмотря на риск переобучения модели, **target encoding** будет более предпочтительным вариантом, если, например, предсказывать цену аппартаментов из-за связи категории с целевой переменной.

```python
df_encoded = df_encoded_target
df_encoded.head()
```

### Генерация новых признаков

Посчитаем цену аренды аппараментов за одну ночь:

```python
df_encoded['price_per_night'] = (df_encoded['price'] / df_encoded['minimum_nights']).round(2)
```

Определим доступность аппартаментов в течении всего года:

```python
df_encoded['availability_ratio'] = (df_encoded['availability_365'] / 365).round(2)
```

Оформим бинарный признак, который показывает, имеет ли владелец сразу несколько объявлений о сдаче аппартаментов в данный момент:

```python
df_encoded['multi_listing_host'] = (df_encoded['calculated_host_listings_count'] > 1).astype(int)
```

### F) Итоговые выводы

#### 1. Что я понял про датасет?

- **Датасет** содержит информацию об объявлениях сервиса Airbnb в Нью-Йорке: *район размещения, тип жилья, цену, минимальное количество ночей, число отзывов и другие характеристики*.

- **Пропуски**: в случае с заголовками объявлений и именами владельцев пропуски представляют собой просто какие-то ошибки парсинга или что-то ещё. А вот пропуски в количестве отзывов - это знак того, что у объявления их вовсе не было, как мы убедились на практике.

- Что насчёт **распределения** признаков: у возможной целевой переменной price наблюдается очень сильный "перекос" в сторону более низких цен и довольно длинный хвост *(максимум - 10000!)*. Похожее скошенное распределение имеют и другие признаки, исходя из гистограмм.

- В данных присутствуют **выбросы**: некоторые объявления имеют аномально высокую цену или очень большое минимальное число ночей.

- Большая часть объявлений сосредоточена в районах **Manhattan** и **Brooklyn**, что отражает туристическую привлекательность этих районов.

- Тип жилья **Entire home/apartment** встречается реже, чем **Private room**, однако его средняя цена значительно выше.

- Географические координаты (**longitude** и **latitude**) показывают высокую плотность объявлений в центральных районах города.

- Между числовыми признаками наблюдается **слабая линейная корреляция**, что говорит о том, что цена зависит скорее от категориальных факторов и сложных взаимодействий признаков.

#### 2. Гипотезы и наблюдения

- **Тип жилья существенно влияет на цену**: объявления типа Entire home/apartment в среднем стоят значительно дороже, чем Private room или Shared room.

- **Район размещения влияет на стоимость жилья**: объявления в Manhattan имеют более высокие цены по сравнению с другими районами.

- **Выбросы** в цене могут быть связаны или с элитной недвижимостью, или ошибками ввода данных, или с долгосрочной арендой.

- **Минимальное количество ночей** может быть связано с типом размещения: квартиры чаще сдаются на более длительный срок.

- **Количество отзывов** может косвенно отражать популярность объявления и потенциально связано с ценой и расположением.

#### 3. Что можно сделать дальше

Выбранный датасет очень хорошо подойдёт для предсказания цены на аренду жилья, а это типичная задача **регрессии**.

После EDA и предобработки датасета, данные следует разбить на **тренировачную** *(обучение модели)*, **валидационную** *(проверка нескольких моделей)* и **тестовую** *(незнакомые данные, с помощью которых оценим конечное качество модели)* выборки.

Дальше - выбор моделей под нашу задачу. Можно было бы взять модель **линейной регрессии** *(как самую базовую)*, **дерево решений** для регресии, а также ансамбли - **случайный лес**, **градиентный бустинг** и/или **XGBoost**.

С помощью **метрик** для регрессии *(RMSE, MAE, R^2)* оценили, насколько хорошие предсказания будут давать выбранные модели и в конечном итоге остановились бы на одной.  

Выбранная модель также смогла бы найти интересные занономерности и связи у признаков, которые мы взяли для обучения.
